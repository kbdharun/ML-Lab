##### Types of Arrays

```python
# 2D Array - (rows,columns)
arr_2 = np.array([[10, 20, 30, 40, 50, 60],
                  [21, 22, 31, 42, 51, 62],
                  [11, 32, 13, 40, 45, 36]])
print(arr_2)
```

```python
print(arr_2.shape)    # return shape = number of elements for 1D
print(arr_2.ndim)     # ndim = returns dimentionality
print(arr_2.size)     # number of items present in an array
```

```python
# 3-D Array - Array of Matrices
arr_3 = np.array([[[10, 20, 30, 40, 50, 60],
                  [21, 22, 31, 42, 51, 62],
                  [11, 32, 13, 40, 45, 36]],

                 [[10, 20, 30, 40, 50, 60],
                  [21, 22, 31, 42, 51, 62],
                  [11, 32, 13, 40, 45, 36]]])
arr_3
```

##### Data Manipulation

```python
# replace = replace ('?') by np.nan
auto_price.replace('?', np.nan, inplace = True)
```

##### Slicing & Indexing DataFrame

```python
# loc = selecting columns & rows by label names
auto_price.loc[1:10,:]
```

```python
# all rows, column labels are used to slice columns
auto_price.loc[:,'make':'engine-size']
```

```python
auto_price.loc[10:20,'make':'height']
```

```python
# slicing using row indexes
auto_price.iloc[1:10,:]
```

```python
auto_price.iloc[1:10,2:12]
```

```python
auto_price.iloc[:,2:12]
```

##### Select Columns

```python
auto_price['price']   # select one column
```

```python
auto_price[['make','body-style','fuel-type','price']]   # select multiple columns
```

##### Filter
    * To filter rows by boolean condition.

```python
auto_price[auto_price['make'] == 'audi']
```

```python
auto_price[auto_price['highway-mpg'] > 30]['price']
```

```python
auto_price[(auto_price['highway-mpg'] > 30) & (auto_price['num-of-doors'] == 'two')]
```

```python
auto_price[(auto_price['body-style'] == 'sedan') | (auto_price['num-of-doors'] == 'two')]
```

```python
# Distinct Values
auto_price['make'].unique()
```

```python
auto_price['make'].value_counts()
```

##### Drop()
    * Dropping Rows & Columns

```python
auto_price.drop(['symboling','normalized-losses'],axis = 1)
```

```python
# Drop Rows
auto_price.drop([2,4,6,8,10,12,14,16],axis = 0)
```

##### Statistical Methods

```python
# typecasting string data into numerical
auto_price['price'] = auto_price['price'].astype('float')
```

```python
# mean()
auto_price['price'].mean()
```

```python
# median()
auto_price['price'].median()
```

```python
# most frequent value - where to apply ? in categorical dataset
auto_price['make'].mode()
```

```python
auto_price['normalized-losses'] = auto_price['normalized-losses'].astype('float')
auto_price['normalized-losses'].fillna(auto_price['normalized-losses'].mean(), inplace = True)
```

```python
auto_price['price'].fillna(auto_price['price'].mean(), inplace = True)
```
