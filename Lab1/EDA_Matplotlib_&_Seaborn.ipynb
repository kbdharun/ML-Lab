```python
import pandas as pd
import numpy as np
# matplotlib - package for basic data viz plot : bar,scatter,line,pie,hist plots
import matplotlib.pyplot as plt
# seaborn package used for descriptive visulaization - KDE(density),boxplot,violin,heatmap plots
import seaborn as sns
```

```python
auto_price = pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/Automobile_price_data__Raw_.csv")
auto_price.head()
```

```python
auto_price.drop(['symboling','normalized-losses'],axis = 1, inplace = True)
```

```python
# replace all columns '?' by np.nan
auto_df = auto_price.replace('?', np.nan)
```

```python
auto_df.dropna(inplace = True)
```

```python
cols = ['bore','stroke','horsepower','peak-rpm','price']
# method to change columns to numeric data types
auto_df[cols] = auto_df[cols].apply(pd.to_numeric)
```

##### Line Plot
    * Univariate & Bivariate (Replationship Plots)
    * Time Series Analysis - Univariate Plots

```python
auto_df['price'].sort_values().unique()
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize = (10,7))
plt.plot(auto_df['price'].sort_values().unique())
plt.show()
```

```python
auto_df.sort_values('price')
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize = (9,6))
plt.plot(auto_df.sort_values('price')['price'],
         auto_df.sort_values('price')['city-mpg'], color = 'b')
plt.title('Bivariate Line Plot - Price vs Mileage', size = 12)
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.show()
```

##### Bar Plot
    * Bivariate / Univariate  Analysis
    * Bar Chart used by Categorical, Nominal Data
    * Use Bar Plot for limited categocal responses.

```python
auto_df['make'].value_counts()
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = auto_df['make'].value_counts().index,
        height = auto_df['make'].value_counts().values)
plt.xticks(rotation = 90)
plt.show()
```

##### Histogram
    * Continious Data - Study spread / distribution of continious data.
    * Univariate Analysis.

```python
plt.figure(figsize = (10,6))
# bins = num of intervals
plt.hist(auto_df['price'], bins = 15)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram Plot - Vehicle Price')
plt.show()
```

##### Scatter PLot
    * Scatter Plots is plotted between numerical values.
    * It shows relationship plots b/w x & y.
    * Bivariate Analysis

```python
plt.figure(figsize = (7,5))
# marker = 's','*','^','>','<','o','.','d'
sns.scatterplot(x = auto_df['city-mpg'], y = auto_df['price'], color = 'deeppink',
                marker = 's', alpha = 0.6)
plt.show()
```

```python
plt.figure(figsize = (7,5))
# marker = 's','*','^','>','<','o','.','d'
# alpha = opacity of color
sns.scatterplot(x = auto_df['horsepower'], y = auto_df['price'], color = 'b',
                marker = 'o', alpha = 0.6)
plt.show()
```

###### Pie
    * Univariate Distribution Analysis.

```python
plt.style.use('ggplot')
plt.figure(figsize = (6,6))
plt.pie(auto_df['make'].value_counts(), autopct = '%0.0f%%',
        labels = auto_df['make'].value_counts().index)
plt.show()
```

##### Kernel Density Estimator (KDE)
    * Distribution Plot : Histogram + Density Plot
    * Probability Density Function (PDF)

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(auto_df['engine-size'], bins = 20, color = 'b')
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
sns.distplot(auto_df['engine-size'], bins = 20, color = 'b', hist = False)
plt.show()
```

###### Correlation
    * Correlation represents linear relationship between two variables.
    * corr() range from -1 to +1, corr b/w two variables = +ve, both are directly proportional
    * corr b/w two variables = -ve, both are inversly proportional
    * corr b/w two variables = 0 or closer to zero, there is no relationalship b/w variables.

```python
# correlation shows linear relationship b/w variables,
auto_df.corr()
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(auto_df.corr(), annot = True)
```

```python
plt.figure(figsize = (3,7))
sns.heatmap(auto_df.corr()[['price']], annot = True)
```

```python
sns.countplot(x = 'body-style', data = auto_df)
```

```python
plt.figure(figsize = (8, 3))
sns.boxplot(auto_df['price'], orient = 'h')
```
